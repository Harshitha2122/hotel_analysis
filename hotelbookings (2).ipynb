```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df=pd.read_csv("C:\\Users\\LENOVO\\Downloads\\archive (2).zip")
print(df.head())
#!ping -n 3 google.com
#!pip install seaborn
print(df.shape)
print(df.columns)
print(df.info)

print(df.isnull().sum())

print("Total bookings:",df.shape[0])#(0 means(not cammceled 1 means canceled number of rows nd 1 means number of columns)
print(df['is_canceled'].value_counts())

print(df['hotel'].value_counts)
sns.countplot(data=df,x='hotel')
plt.title("Hotel type Bookings")
plt.show()

print(df['customer_type'].value_counts())
print(df['deposit_type'].value_counts())
print(df['arrival_date_day_of_month'].value_counts())
print(df['reservation_status'].value_counts())

sns.countplot(data=df,x='is_canceled')
plt.title("Canceled vs Not Canceled Bookings")
plt.xticks([0,1],['Not Canceled','Canceled'])
plt.show()

sns.countplot(data=df,x='arrival_date_month', order=['January', 'February',  'March', 'April', 'May', 'June', 'July',  'September', 'October', 'November', 'December'])
plt.title("Monthly Bookings")
plt.xticks(rotation=45)
plt.show()

df.to_csv('hotel_bookings.csv',index=False)
```
